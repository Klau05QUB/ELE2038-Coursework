##Linearisation

Linearisation of $\overline{x_2}$

```python
#Numerical values to use in substitutions

m_val = 0.462
g_val = 9.81
d_val = 0.42
delta_val = 0.65
r_val = 0.123
R_val = 2200
L0_val = 0.125
L1_val = 0.0241
alpha_val = 1.2
c_val = 6.811
k_val = 1885
b_val = 10.4
tau_val = 0.03
phi_val = np.deg2rad(41)

x_1e_val = 0.5
x_2e_val = 0
x_3e_val = (delta_val - x_1e_val)*sqrt( (k_val*(x_1e_val-d_val) - m_val*g_val*sin(phi_val)) / c_val)

V_e_val = x_3e_val * R_val

```

```python
delta, K, m, b, c = sym.symbols('delta, K, m, b, c')
x1, x_1e = sym.symbols('x1, x_1e')
x2 = sym.symbols('x2')
x3, x_3e = sym.symbols('x3, x_3e')
x1bar, x2bar, x3bar = sym.symbols(r'\bar{x_1} \bar{x_2} \bar{x_3}')

#Gamma is the nonlinear term to be linearised

gamma = x3**2*c/((delta-x1)**2) - x_3e**2*c/((delta-x_1e)**2)
gamma
```

```python
d_gamma_x1 = gamma.diff(x1).simplify()
d_gamma_x1 = d_gamma_x1.subs([ (x1,x_1e) , (x3,x_3e) , (x2,0) ]).simplify()

gamma_x1_coefficient = (5/(7*m))*(d_gamma_x1 - K)

gamma_x1_coefficient = gamma_x1_coefficient.subs([ (x_1e,x_1e_val) , (x_3e,x_3e_val) , (x2, x_2e_val) , (m, m_val), (b, b_val), (K, k_val), (delta, delta_val), (c, c_val)]).simplify()


d_gamma_x3 = gamma.diff(x3).simplify()
d_gamma_x3 = d_gamma_x3.subs([ (x1,x_1e) , (x3,x_3e) , (x2,0) ]).simplify()

gamma_x3_coefficient = (5/(7*m))*(d_gamma_x3)
gamma_x3_coefficient = gamma_x3_coefficient.subs([ (x_1e,x_1e_val) , (x_3e,x_3e_val) , (x2, x_2e_val) , (m, m_val), (b, b_val), (K, k_val), (delta, delta_val), (c, c_val)]).simplify()


gamma_x2_coefficient = (5/(7*m))*(-b)
gamma_x2_coefficient = gamma_x2_coefficient.subs([ (x_1e,x_1e_val) , (x_3e,x_3e_val) , (x2, x_2e_val) , (m, m_val), (b, b_val), (K, k_val), (delta, delta_val), (c, c_val)]).simplify()

print(gamma_x1_coefficient)
print(gamma_x3_coefficient)
print(gamma_x2_coefficient)
```

```python
dot_x2_bar_linearised = gamma_x1_coefficient*x1bar + gamma_x3_coefficient*x3bar + gamma_x2_coefficient*x2bar

dot_x2_bar_linearised
```

Linearisation of $\overline{x_3}$

```python
from sympy.functions import exp

#Declaration of symbols used

delta, alpha,  R, L0, L1 = sym.symbols('delta, alpha, R, L0, L1')
x1, x_1e = sym.symbols('x1, x_1e')
x2 = sym.symbols('x2')
x3, x_3e = sym.symbols('x3, x_3e')
V, V_e = sym.symbols('V, V_e')
x1bar, x2bar, x3bar, Vbar = sym.symbols(r'\bar{x_1} \bar{x_2} \bar{x_3} \bar{V}')

zeta = (V - x3*R)/(L0+L1*exp(-alpha*(delta - x1)))
zeta
```

```python
d_zeta_V = zeta.diff(V).simplify()
d_zeta_V = d_zeta_V.subs([ (x1,x_1e) , (V,V_e) , (V,V_e) ]).simplify()
VCoefficient = d_zeta_V.subs([ (x_1e,x_1e_val) , (x_3e,x_3e_val) , (x2, x_2e_val) , (m, m_val), (b, b_val), (K, k_val), (delta, delta_val), (c, c_val), (V_e, V_e_val)]).simplify()


d_zeta_x3 = zeta.diff(x3).simplify()
d_zeta_x3 = d_zeta_x3.subs([ (x1,x_1e) , (x3,x_3e) , (V,V_e) ]).simplify()
zeta_x3_coefficient = d_zeta_x3.subs([ (x_1e,x_1e_val) , (x_3e,x_3e_val) , (x2, x_2e_val) , (m, m_val), (b, b_val), (K, k_val), (delta, delta_val), (c, c_val), (V_e, V_e_val)]).simplify()

d_zeta_x1 = zeta.diff(x1).simplify()
d_zeta_x1 = d_zeta_x1.subs([ (x1,x_1e) , (x3,x_3e) , (V,V_e) ]).simplify()
zeta_x1_coefficient = d_zeta_x1.subs([ (x_1e,x_1e_val) , (x_3e,x_3e_val) , (x2, x_2e_val) , (m, m_val), (b, b_val), (K, k_val), (delta, delta_val), (c, c_val), (V_e, V_e_val)]).simplify()
```

```python
dot_x_3_bar_linearised = d_zeta_x1*x1bar + VCoefficient*Vbar + zeta_x3_coefficient*x3bar
dot_x_3_bar_linearised = dot_x_3_bar_linearised.subs([ (x_1e,0.5) , (x_3e,0.7) , (x2,0), (delta, 0.65), (L1, 0.0241), (L0, 0.125), (alpha, 1.2), (R, 2200), (V_e, 0.7*2200)]).simplify()
dot_x_3_bar_linearised
```
